```python
def plotDiff(data1, data2):
    diffs = [point[0] - point[1] for point in zip(data1, data2)]
    diffs_mean = np.mean(diffs)
    print('average diff: {}'.format(diffs_mean))
    diffs_std = np.std(diffs)
    diffs_normalized = (diffs - diffs_mean) / diffs_std
    fig, ax = plt.subplots()
    ax.plot(diffs_normalized)
    
def plotRatio(data1, data2):
    diffs = [point[0] / point[1] for point in zip(data1, data2)]
    diffs_mean = np.mean(diffs)
    print('average ratio: {}'.format(diffs_mean))
    diffs_std = np.std(diffs)
    diffs_normalized = (diffs - diffs_mean) / diffs_std
    plt.plot(diffs_normalized)
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

def bollinger_lower(a, n=3):
    cleaned = moving_average(a, n)
    std = np.std(a)
    return cleaned - std
```
